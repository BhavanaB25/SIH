```python
import os
import nibabel as nib
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Define input & output paths
base_path = "/content/drive/MyDrive/MyProject/DTI/DTI_Categorized"
output_path = "/content/drive/MyDrive/MyProject/DTI/DTI_Preprocessed"

# Ensure output directory exists
os.makedirs(output_path, exist_ok=True)

# Resize shape (modify based on your model requirements)
IMG_SIZE = (128, 128)

# Function to preprocess DTI images
def preprocess_dti(input_folder, output_folder):
    """Preprocess DTI images: load, resize, normalize, and save."""
    for category in ["AD", "CN", "MCI"]:
        input_category_path = os.path.join(input_folder, category)
        output_category_path = os.path.join(output_folder, category)
        os.makedirs(output_category_path, exist_ok=True)

        for patient_id in os.listdir(input_category_path):
            patient_folder = os.path.join(input_category_path, patient_id)

            # Check if the patient folder exists
            if not os.path.isdir(patient_folder):
                continue

            for file in os.listdir(patient_folder):
                if file.endswith(".nii") or file.endswith(".nii.gz"):  # Check if it's a NIfTI file
                    file_path = os.path.join(patient_folder, file)

                    try:
                        # Load NIfTI image
                        img = nib.load(file_path).get_fdata()

                        # Normalize image (0 to 1 range)
                        img = (img - np.min(img)) / (np.max(img) - np.min(img) + 1e-8)

                        # Resize slices
                        resized_slices = []
                        for i in range(img.shape[2]):  # Iterate over slices
                            slice_resized = cv2.resize(img[:, :, i], IMG_SIZE, interpolation=cv2.INTER_AREA)
                            resized_slices.append(slice_resized)

                        resized_slices = np.array(resized_slices)

                        # Save preprocessed image as .npy file
                        output_file = os.path.join(output_category_path, f"{patient_id}.npy")
                        np.save(output_file, resized_slices)

                        print(f"‚úÖ Processed: {file_path} ‚Üí {output_file}")

                    except Exception as e:
                        print(f"‚ùå Error processing {file_path}: {e}")

# Run preprocessing
preprocess_dti(base_path, output_path)
print("\nüéâ Preprocessing Complete! Preprocessed data saved in:", output_path)
```

```python
from monai.networks.nets import EfficientNetBN
import torch
import torch.nn as nn
from monai.networks.nets import resnet, densenet

class DTI_Classifier(nn.Module):
    def __init__(self, num_classes=3):
        super(DTI_Classifier, self).__init__()

        # 3D ResNet
        self.resnet3d = resnet.resnet50(spatial_dims=3, n_input_channels=1, num_classes=num_classes)

        # 3D DenseNet
        self.densenet3d = densenet.densenet121(spatial_dims=3, in_channels=1, out_channels=num_classes)

        # 3D EfficientNet
        self.efficientnet3d = EfficientNetBN(model_name="efficientnet-b0", in_channels=1, num_classes=num_classes)

        # Fully connected layer to combine outputs from all models
        self.fc = nn.Linear(num_classes * 3, num_classes)

    def forward(self, x):
        res_out = self.resnet3d(x)
        dense_out = self.densenet3d(x)
        eff_out = self.efficientnet3d(x)

        # Concatenating outputs from all networks
        combined_features = torch.cat((res_out, dense_out, eff_out), dim=1)

        # Final classification
        out = self.fc(combined_features)
        return out

# Define model
model = DTI_Classifier(num_classes=3)

# Print model summary
print(model)
```

```python
# Define loss and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Move model to GPU if available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
```

```python
def evaluate_model(model, test_loader):
    model.eval()
    correct, total = 0, 0

    with torch.no_grad():
        for images, labels in test_loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            _, predicted = torch.max(outputs, 1)
            correct += (predicted == labels).sum().item()
            total += labels.size(0)

    acc = correct / total
    print(f"Test Accuracy: {acc:.4f}")

# Call function
# evaluate_model(model, test_loader)
```
